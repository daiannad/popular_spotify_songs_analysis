```python
# line plot to show number of songs released each year 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import EventCollection
song_counts_by_year = df['released_year'].value_counts().sort_index()

xdata = song_counts_by_year.index
ydata = song_counts_by_year.values
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(xdata, ydata, color='tab:blue')
xevents = EventCollection(xdata, color='tab:blue', linelength=0.1)

yevents = EventCollection(ydata, color='tab:blue', linelength=0.1, orientation='vertical')

ax.add_collection(xevents)
ax.add_collection(yevents)

ax.set_xlim([xdata.min(), xdata.max()])
ax.set_ylim([ydata.min(), ydata.max()])

ax.set_title('Number of Songs Released Each Year')
plt.show()
```

```python
#horizontal bar chart to show number of songs in spotify and apple playlists
import matplotlib.pyplot as plt
import numpy as np

playlists = ('Spotify Playlist', 'Apple Playlist')
playlist_counts = np.array([100, 120])  # Example counts for Spotify and Apple playlists
error = np.random.rand(len(playlists))

fig, ax = plt.subplots()
ax.barh(playlists, playlist_counts, xerr=error, align='center', color=['blue', 'green'])

ax.set_xlabel('Number of Songs')
ax.set_title('Number of Songs in Spotify and Apple Playlists')

plt.show()
```
